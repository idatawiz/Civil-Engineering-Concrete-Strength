```python
Target='Strength'
Predictors=['CementComponent ','SuperplasticizerComponent','AgeInDays']
```

# Multiple Linear Regression

```python
 from sklearn.linear_model import LinearRegression
```

```python
RegModel=LinearRegression()
```

```python
print(RegModel)
```

```python
LREG=RegModel.fit(X_train,Y_train)
from sklearn import metrics
print('R Squared Value is:',metrics.r2_score(Y_train,LREG.predict(X_train)))
```

```python
print('Model Validation & Accuracy')
prediction=LREG.predict(X_test)
TDR=pd.DataFrame(data=X_test,columns=Predictors)
TDR[Target]=Y_test
TDR[('Predicted'+Target)]=np.round(prediction)

print('calculating the error of prediction for each row')
TDR['APE']=100*((abs(TDR['Strength']-TDR['PredictedStrength']))/TDR['Strength'])
print(TDR[[Target,'Predicted'+Target,'APE']].head())

MeanAPE=np.mean(TDR['APE'])
MedianAPE=np.median(TDR['APE'])

ACC=100-MeanAPE
MACC=100-MedianAPE
print('Mean Accuracy on test data',ACC)
print('Median Accruracy on Test data',MACC)

def acc_score(orgi,predic):
    Mape=np.mean(100*(np.abs(orgi-predic)/orgi))
    return(100-Mape)
from sklearn.metrics import make_scorer
custom_scoring=make_scorer(acc_score,greater_is_better=True)

from sklearn.model_selection import cross_val_score
Acc_values=cross_val_score(RegModel,X,Y,cv=10,scoring=custom_scoring)
print('Accuracy values for 10 cross fold validation',Acc_values)
print('Final Avg Model Accuracy',round(Acc_values.mean(),2))
```

```python
def FunctionPredictResult(InputData):
    import pandas as pd
    Num_Inputs=InputData.shape[0]
    
    DataForML=pd.read_pickle('DataForMLConcS.pkl')
    InputData=InputData.append(DataForML)
    
    Predictors=['CementComponent ','SuperplasticizerComponent','AgeInDays']
    X=InputData[Predictors].values[0:Num_Inputs]
    
    X=PredictorScalerFit.transform(X)
    
    
    import pickle
    with open('Final_concsRF_Model.pkl', 'rb') as fileReadStream:
        PredictionModel=pickle.load(fileReadStream)
        
        fileReadStream.close()
            

    Prediction=PredictionModel.predict(X)
    PredictionResult=pd.DataFrame(Prediction, columns=['Prediction'])
    return(PredictionResult)
```

```python
def FunctionGeneratePrediction(inp_CementComponent, inp_SuperplasticizerComponent, inp_AgeInDays):
    SampleInputData=pd.DataFrame(
     data=[[inp_CementComponent, inp_SuperplasticizerComponent, inp_AgeInDays]],
     columns=['CementComponent','SuperplasticizerComponent','AgeInDays'])
    Predictions=FunctionPredictResult(InputData= SampleInputData)
    
    return(Predictions.to_json())
FunctionGeneratePrediction(inp_CementComponent=800, 
                           inp_SuperplasticizerComponent=5.96, 
                           
                           inp_AgeInDays=14
                             )
```

```python
app = Flask(__name__)

@app.route('/my_prediction', methods=["GET"])
def prediction_api():
    try:
        CementComponent_value=float(request.args.get('CementComponent'))
        SuperplasticizerComponent_value=float(request.args.get('SuperplasticizerComponent'))
        AgeInDays_value=float(request.args.get('AgeInDays'))
        
                       
        prediction_from_api=FunctionGeneratePrediction(
                                                    inp_CementComponent=CementComponent_value, 
                                                    inp_SuperplasticizerComponent=SuperplasticizerComponent_value, 
                                                    inp_AgeInDays=AgeInDays_value
                                                                                                       

                                                        )

        return (prediction_from_api)
    
    except Exception as e:
        return('Something is not right!:'+str(e))
```
